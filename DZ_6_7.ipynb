Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки

Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression из модуля sklearn.linear_model.\
Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.\

```python
# Импортируйте библиотеки pandas и numpy.
import pandas as pd
import numpy as np
```

```python
# Загрузите "Boston House Prices dataset" из встроенных наборов данных библиотеки sklearn.
from sklearn.datasets import load_boston
boston = load_boston()
boston.keys()
```

```python
# подготовим массив данных для датафреймов переменной X
data = boston.data
data
```

```python
# посмотрим размерность массива
data.shape
```

```python
# подготовим массив данных для датафреймов переменной y
target = boston.target
target
```

```python
# перенесём имена колонок из старого в новый датафрейм
feature_names = boston.feature_names
feature_names
```

```python
# Создайте датафрейм X из этих данных.
X = pd.DataFrame(data, columns=feature_names)
X.head()
```

```python
# информация по датафрейму X
X.info()
```

```python
# Создайте датафрейм y из этих данных.
y = pd.DataFrame(target, columns=['price'])
y.head()
```

```python
# информация по датафрейму y
y.info()
```

```python
# загрузим метод подготовки тренировочных данных из библиотеки
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# загрузим метод линейной регрессии из библиотеки
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# запуск функции обучения
lr.fit(X_train, y_train)
```

```python
# создаём переменную для сравнения с реальными значениями
y_pred = lr.predict(X_test)
y_pred.shape
```

```python
# сравниваем переменные
check_test = pd.DataFrame({
 "y_test": y_test["price"],
 "y_pred": y_pred.flatten(),
})
check_test.head(10)
```

```python
# проверка величины ошибки
check_test["error"] = check_test["y_pred"] - check_test["y_test"]
check_test.head()
```

```python
# 
from sklearn.metrics import r2_score
r2_score_1=r2_score(check_test["y_pred"], check_test["y_test"])
r2_score_1
```

```python
# Создайте модель под названием model с помощью RandomForestRegressor из модуля sklearn.ensemble.
# Сделайте агрумент n_estimators равным 1000, max_depth должен быть равен 12 и random_state сделайте равным 42.
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
```

```python
# Обучите модель на тренировочных данных аналогично тому, как вы обучали модель LinearRegression,
# но при этом в метод fit вместо датафрейма y_train поставьте y_train.values[:, 0], 
# чтобы получить из датафрейма одномерный массив Numpy, так как для класса RandomForestRegressor
# в данном методе для аргумента y предпочтительно применение массивов вместо датафрейма.

model.fit(X_train, y_train.values[:, 0])
```

```python
# предсказательная модель
y_pred = model.predict(X_test)
y_pred.shape
```

```python
# сравнение результатов
check_test = pd.DataFrame({
 "y_test": y_test["price"],
 "y_pred": y_pred.flatten(),
})
check_test.head(10)
```
